```python
# ### Creating Cohorts based on Tenure

# **Let's begin by treating each unique tenure length, 1 month, 2 month, 3 month...N months as its own cohort.**

# **TASK: Treating each unique tenure group as a cohort, calculate the Churn rate (percentage that had Yes Churn) per cohort. 
#     For example, the cohort that has had a tenure of 1 month should have a Churn rate of 61.99%. You should have cohorts 1-72 
#     months with a general trend of the longer the tenure of the cohort, the less of a churn rate. This makes sense as you are 
#     less likely to stop service the longer you've had it.**
```

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
yes_churn = df.groupby(['Churn', 'tenure']).count().transpose()['Yes']
```

```python
yes_churn
```

```python
churn_rate = (100 * yes_churn) / (no_churn + yes_churn)
```

```python
churn_rate.transpose()['customerID']
```

```python
#TASK: Now that you have Churn Rate per tenure group 1-72 months, create a plot showing churn rate per months of tenure.
```

```python
plt.figure(figsize = (10,4), dpi = 200)
churn_rate.iloc[0].plot()
plt.ylabel('Churn_%ge');
```

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return 'Over 48 Months'
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['Tenure Cohort', 'tenure']]
```

```python
X = df.drop(['customerID', 'Churn'], axis = 1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
preds = dt.predict(X_test)
```

```python
print(classification_report(y_test, preds))
```

```python
plot_confusion_matrix(dt, X_test, y_test)
```

```python
dt.feature_importances_
```

```python
imp_feats = pd.DataFrame(data = dt.feature_importances_, index = X.columns, columns=['Feature Importances']).sort_values('Feature Importances')
```

```python
imp_feats = imp_feats[imp_feats['Feature Importances']>0]
```

```python
imp_feats
```

```python
plt.figure(figsize = (10,4), dpi = 200)
sns.barplot(data = imp_feats, x = imp_feats.index, y = 'Feature Importances')
plt.xticks(rotation = 90);
plt.title('Feature Importances Of Decision Tree')
```
